```python
import numpy as np
from tensorflow import keras
```

# Define and Compile Neural Network

Create simplest neural network. It has 1 layer with 1 neuron and input shape to it is just 1 value 

```python
#Build a simple Sequential model
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

Here we will train model to idententify relation between numbers y=2x-1.
When tring to learn model will make a guess. 
The loss function measures guessed anwer against knowm answer and measure how well or how badly it did.
The optimizer function make another guess ad ==nd tries to minimize loss

To calculate loss use mean squared error and for optimizer use stochastic gradient descent(opyimization algorithm)

```python
#Compile the model
model.compile(optimizer='sgd',loss='mean_squared_error')
```

Now feed in some data

```python
#declare models inputs and outputs for training
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype = float)
```

# Training Neural Network

epochs are number of time model will go through the loop of loos and optimization

```python
#Train the model
model.fit(xs, ys, epochs=500)
```

Let's predict Y for x = 10 (correct anwer 19)

```python
# Convert input data to a NumPy array
input_data = np.array([10.0])

# Use the model to make predictions
predictions = model.predict(input_data)

print(predictions)
```
